# Air Quality Data Study 

## Overview 
The following project explores air quality trends and variations for the city of Islamabad over a period of 3.5 years. 

### Scientific information: 
What is Air Quality?
Air quality refers to how clean or polluted the air. It depends on the number of pollutants found in air. Various and common substances that affect air quality include: 
1. Particulate Matter such as PM2.5\
   Includes tiny inhable particles such as dust, smoke and chemicals. '2.5' refers to the diameter size of particles
2. Sulphur Dioxide (SO_2)\
    It is a nonflammable gas that is soluble in water, produced by burning fossil fuels and industrial processes
3. Nitrogen Dioxide (NO_2)\
   It is a pungent, acidic, and highly oxidizing gas, produced by of burning of fuel in automobiles and power plants

Affect on health and environment:\
High concentrations and long term exposure to these particles causes severe respiratory and cardiac discorders. Affects environment by reducing visibility, forming acid rains and overall impacts the ecosystem and it's habitants alike. 

## Objectives 
The purpose of this study is to explore the concentration trends of various air pollutants such as PM2.5, NO2, SO2 and affects of certain environemntal factors such as temperature and humidity on PM2.5 concentration. 

### Questions addressed: 
1. Variation in concentrations of NO2, SO2 and PM2.5, monthly over a period of 3 to 4 years.
2. Variation in concentration levels seasonally
3. Affect of NO2, SO2, temperature and humidity on PM2.5 concentration levels 

## Data
The data consists of corresponding values of temperature, humidity, NO2, SO2 and PM2.5 concentration recorded between June 2019 to Mar 2023 for the city of Islamabad, Pakistan. Tempertaure has been recorded in C, Humidity in %, while NO2, SO2 and PM2.5 concentrations in 'µg/cm³'. The data has been obtained from Kaggle.com (https://www.kaggle.com/datasets/diraf0/islamabad-pakistan-air-quality-data), as indicated by publisher, is officially available on Pakistan Environmental Protection Agency (EPA) website. The publisher has distributed the data under 'Public Domain'.  

## Methodology
The following study has been done by adopting various data analysis techniques:
1. Monthly, yearly and seasonal trends:\
   Scatter and line plots to study pollutant concentration trends.
2. Relation between environmental factors such as temperature and humidity with pollutant concentrations:\
   Linear regression fitting and correlation analysis.
5. Predicting PM2.5 values
   By training and implementing regression models

```python
import pandas as pd
from matplotlib import pyplot as plt
from scipy import stats

data_temp_vs_pm25 = {
    'temp' : [27.54,30.73,29.30,29.21,25.80,17.83,13.37,7.36,16.61,17.15,19.50,26.66,31.37,32.55,32.85,
            31.87,26.02,17.55,10.77,8.20,14.13,21.08,23.28,25.58,31.77,32.71,27.05,27.43,24.20,16.45,11.03,13.74,12.32,
            21.41,28.17,32.71,33.60,29.50,28.70,28.44,23.57,25.50,13.38],
    'pm25' : [29.02, 32.15, 28.05, 36.40, 50.95, 49.65, 62.71,44.04,56.59,25.84,17.67,20.63,19.55,27.15,
            23.08,24.45,34.21,39.90,51.11,53.91,50.33,21.68,16.32,19.37,20.33,17.35,23.81,31.48,20.98,53.29,104.77,
           77.40,58.51,39.20,24.03,22.37,25.16,17.71,19.97,22.26,25.62,19.39,65.46]  
}

df_temp_vs_pm25 = pd.DataFrame(data_temp_vs_pm25)

#apply regression fit

df_temp_vs_pm25_reg = df_temp_vs_pm25[['temp', 'pm25']].copy()

# Get the regression slope and intercept
m, b, r, p, se = stats.linregress(df_temp_vs_pm25_reg['temp'], df_temp_vs_pm25_reg['pm25'])
print('slope: {:.4f}\ny-intercept: {:.4f}'.format(m,b))
print('so...\n f(x) = {:.4f}x + {:.4f}'.format(m,b))

# Use the function (mx + b) to calculate f(x) for each x  value
df_temp_vs_pm25_reg['fx'] = (m * df_temp_vs_pm25_reg['temp']) + b

# Calculate the error between f(x) and the actual y value
df_temp_vs_pm25_reg['error'] = df_temp_vs_pm25_reg['fx'] - df_temp_vs_pm25_reg['pm25']

# Create a scatter plot 
df_temp_vs_pm25_reg.plot.scatter(x='temp', y='pm25',color = 'purple')
plt.title('Temperature vs PM2.5 comparison (June 19 - Dec 22)')
plt.ylabel('PM2.5 concentration (µg/cm³)')
plt.xlabel('Temperature (°C)')
plt.grid(True)

# Plot the regression line
plt.plot(df_temp_vs_pm25_reg['temp'],df_temp_vs_pm25_reg['fx'], color='black')

# Display the plot
plt.show()
```

## 9. Prediction of PM2.5 particle concentration using ML models

For this part of analysis, regression models were fitted on the entire data to help evaluate linear relation between the predictors such as temperature, humidity, NO2 and SO2 and the target variable which is PM2.5. The goal was to predict whether the data for predictors can be used to find approximate values of PM2.5 concentration.\
\
For analysis from here on, the entire data has been handles using a much simpler approach. All values of predictors and PM2.5 were concatinated into a single dataframe. The model was trained on 70% of the data while remaining 30% was used to test the model. 

```python
# Make single data frame 

all_data = [data19, data20, data21, data22, data23]

var_temp = []

for yearly_dict in all_data:
    for df in yearly_dict.values():
        var_temp.extend(df["Temperature"].dropna().tolist())

var_hum = []

for yearly_dict in all_data:
    for df in yearly_dict.values():
        var_hum.extend(df["Humidity"].dropna().tolist())

var_NO2 = []

for yearly_dict in all_data:
    for df in yearly_dict.values():
        var_NO2.extend(df["NO2"].dropna().tolist())

var_SO2 = []

for yearly_dict in all_data:
    for df in yearly_dict.values():
        var_SO2.extend(df["SO2"].dropna().tolist())

var_pm25 = []

for yearly_dict in all_data:
    for df in yearly_dict.values():
        var_pm25.extend(df["PM2.5"].dropna().tolist())


combined_data = pd.concat(
    df[["Temperature", "Humidity", "NO2", "SO2", "PM2.5"]]
    for yearly_dict in all_data
    for df in yearly_dict.values()
)

combined_data = combined_data.dropna()

print(combined_data)
```

```python
# Separate features and tareget variable

X, y = combined_data[['Temperature', 'Humidity', 'NO2', 'SO2']].values, combined_data['PM2.5'].values
print(X,y, sep='\n')
```

```python
from sklearn.model_selection import train_test_split

# Split data 70%-30% into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)

print ('Training Set: %d rows\nTest Set: %d rows' % (X_train.shape[0], X_test.shape[0]))
```

```python
# Train the model
from sklearn.linear_model import LinearRegression

# Fit a linear regression model on the training set
model = LinearRegression().fit(X_train, y_train)
print (model)
```

```python
import numpy as np

predictions = model.predict(X_test)
np.set_printoptions(suppress=True)
print('Predicted values: ', np.round(predictions)[:10])
print('Actual values   : ' ,y_test[:10])
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, predictions, color = 'purple')
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.title('PM2.5 predictions')
# overlay the regression line
z = np.polyfit(y_test, predictions, 1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test), color='black')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score

r2 = r2_score(y_test, predictions)
print("R2:", r2)
```

### 9.1 Discussion

To test how well the model has predicted the actual values of PM2.5, the root mean square was computed. 
1. A value of 0.34 indicate that about 34% of the variation in PM2.5 is explained by temperature, NO₂, SO₂ and humidity.
2. Shows moderate explanatory power

Possible factors affecting results:\
Since the data is environmental, many other related predictors such as wind speed and direction, atmospheric pressure, rainfall cannot be ignored while making any conclusions about the given data. 

To test whether better predictions can be made, the random forest model was trained in the similar way. This model does not automatically predict linear relation between predictors and target variable which could be used to improve predictions. 

### 9.2 Train Random Forest algorithm

```python
from sklearn.ensemble import RandomForestRegressor

# Train the model
model = RandomForestRegressor().fit(X_train, y_train)
print (model, "\n")

# Evaluate the model using the test data
predictions = model.predict(X_test)
mse = mean_squared_error(y_test, predictions)
print("MSE:", mse)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
r2 = r2_score(y_test, predictions)
print("R2:", r2)

# Plot predicted vs actual
plt.scatter(y_test, predictions, color = 'orange')
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.title('PM2.5 Predictions')
# overlay the regression line
z = np.polyfit(y_test, predictions, 1)
p = np.poly1d(z)
plt.plot(y_test,p(y_test), color='black')
plt.show()
```

The results have significantly increased from 34% to 52% accuracy. This indicates that trends in data are not completely linear rather cross non-linear relations between predictors and target variable is are present. For example: humidity might increase PM2.5 at low temperatures but decrease it at high temperatures, while linear regression caanot capture such trends a Random Forest model proves a better choice.
